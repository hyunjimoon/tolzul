```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Problem 2

x = np.arange(0, 20, 0.1)
y2 = x + 3 # min
y3 = 9 - 3*x # max
y4 = 1/3*x # max
y5 = 1/2*(x-2) # max

c = 27/10
yo = -4/6*x + c

fig, ax = plt.subplots()

plt.plot(x, y2, color = 'b', label = "Constraint 2")
plt.plot(x, y3, color = 'g', label = "Constraint 3")
plt.plot(x, y4, color = 'c', label = "Constraint 4")
plt.plot(x, y5, color = 'k', label = "Constraint 5")
plt.plot(x, yo, color = 'r', label = "Objective")

# view control
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 20)
plt.ylim(0, 20)
plt.xticks(np.append(20, np.arange(0, 20, 2)))
plt.yticks(np.append(20, np.arange(0, 20, 2)))


yc = np.maximum(np.maximum(y3, y4), y5)
plt.fill_between(x, y2, yc, where=y2>=yc, color='grey', alpha=0.4,
                interpolate=True, label='Feasible Region')
plt.legend(loc='upper left')

# # plot decoration
# from matplotlib.patches import Patch
# # dum_plot = Patch(facecolor='orange', edgecolor='r', label='Color Patch')
# leg = ax.legend(loc='upper left')

# # Annotation
plt.annotate(text='(2.7, 0.9)',xy=(2.7, 0.9), xytext=(4, 3.5),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
plt.annotate(text='(1.5, 4.5)',xy=(1.5, 4.5), xytext=(1, 8),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
plt.show()
```
