```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Problem 1

cycle_length = 60
red = 30
reds = np.array([0, 60, 120])
va = np.array([[0, 0], [60, 1200], [90, 1200], [105, 1500]])
vb = np.array([[20, 0], [57.7, 1500]])

fig, ax = plt.subplots()

# add background color band
for r in reds:
    g = r + red
    ax.axvspan(r, g, facecolor='r', alpha = 0.2)
    ax.axvspan(g, g + cycle_length - red, facecolor='g', alpha=0.2)

# add x-t diagram for vehicles A and B
va = ax.plot(va[:, 0], va[:, 1], color = 'b', label = "Vehicle A")
vb = ax.plot(vb[:, 0], vb[:, 1], color='m', label = "Vehicle B")

ax.axhline(600, 0, 105, color='grey', alpha=0.3)
ax.axhline(1200, 0, 105, color='grey', alpha=0.3)
# view control
ax.set_xlabel("Time (sec)")
ax.set_ylabel("Distance (m)")
ax.set_xticks(np.append(105, np.arange(0, 105, 15)))
ax.set_xlim([0, 105])
ax.set_ylim([0, 1500])

# plot decoration
leg = ax.legend(loc='upper left')

# Annotation
ax.annotate(text='intersection 1',xy=(0, 600), xytext=(1, 500),
            color='grey')
ax.annotate(text='intersection 2',xy=(0, 1200), xytext=(1, 1100),
            color='grey')
ax.annotate(text='(30, 600)',xy=(30, 600), xytext=(15, 700),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(60, 1200)',xy=(60, 1200), xytext=(60, 1300),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(90, 1200)',xy=(90, 1200), xytext=(80, 1000),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(35, 600)',xy=(35, 600), xytext=(40, 500),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(40, 800)',xy=(40, 800), xytext=(45, 800),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(105, 1500)',xy=(105, 1500), xytext=(110, 1400),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(57.5, 1500)',xy=(57.5, 1500), xytext=(30, 1350),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(50, 1200)',xy=(50, 1200), xytext=(27, 1050),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
```

```python
# Problem 2

cycle_length = 2
red = 1
reds = np.array([0, 2])

xa_arr = np.array([[0, 0], [4, 80]])
xa_dep = np.array([[0, 0], [1, 0], [1.5, 30], [2, 40], [3, 40], [3.5, 70]])

fig, ax = plt.subplots()

# add background color band
for r in reds:
    g = r + red
    ax.axvspan(r, g, facecolor='r', alpha = 0.2)
    ax.axvspan(g, g + cycle_length - red, facecolor='g', alpha=0.2)

# add x-t diagram for vehicles A and B
xa_arr = ax.plot(xa_arr[:, 0], xa_arr[:, 1], color = 'b',
                 label = "Road A arrival")
xa_dep = ax.plot(xa_dep[:, 0], xa_dep[:, 1], color = 'm',
                 label = "Road A departure")

# view control
ax.set_xlabel("Time (min)")
ax.set_ylabel("# of Cars")
ax.set_xticks(np.append(4, np.arange(0, 4, 1)))
ax.set_xlim([0, 4])
ax.set_ylim([0, 120])
x = np.array([0, 1, 1.5, 2, 3, 3.5])
y1 = np.array([0, 0, 30, 40, 40, 70])
y2 = np.array([0, 20, 30, 40, 60, 70])
fill = ax.fill_between(x, y1, y2, color='grey', alpha=0.4,
                interpolate=True, label='Road A Total delay')
# plot decoration
from matplotlib.patches import Patch
# dum_plot = Patch(facecolor='orange', edgecolor='r', label='Color Patch')
leg = ax.legend(loc='upper left')

# Annotation
ax.annotate(text='(1.5, 30)',xy=(1.5, 30), xytext=(1.0, 40),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(2, 40)',xy=(2, 40), xytext=(2, 25),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(3, 40)',xy=(3, 40), xytext=(3, 25),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(3.5, 70)',xy=(3.5, 70), xytext=(3, 80),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
```

```python
# Problem 2

cycle_length = 2
red = 1
reds = np.array([1, 3])

xa_arr = np.array([[0, 0], [4, 120]])
xa_dep = np.array([[0, 0], [1, 30], [2, 30], [3, 90]])

fig, ax = plt.subplots()

# add background color band
for r in reds:
    g = r - red
    ax.axvspan(r, r + red, facecolor='r', alpha = 0.2)
    ax.axvspan(g, r, facecolor='g', alpha=0.2)

# add x-t diagram for vehicles A and B
xa_arr = ax.plot(xa_arr[:, 0], xa_arr[:, 1], color = 'b',
                 label = "Road B arrival")
xa_dep = ax.plot(xa_dep[:, 0], xa_dep[:, 1], color = 'm',
                 label = "Road B departure")

# view control
ax.set_xlabel("Time (min)")
ax.set_ylabel("# of Cars")
ax.set_xticks(np.append(4, np.arange(0, 4, 1)))
ax.set_xlim([0, 4])
ax.set_ylim([0, 120])
x = np.array([1, 2, 3])
y1 = np.array([30, 30, 90])
y2 = np.array([30, 60, 90])
fill = ax.fill_between(x, y1, y2, color='grey', alpha=0.4,
                interpolate=True, label='Road B Total delay')
# plot decoration
leg = ax.legend(loc='upper left')

# Annotation
ax.annotate(text='(1, 30)',xy=(1, 30), xytext=(0.5, 40),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(2, 30)',xy=(2, 30), xytext=(2, 15),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
ax.annotate(text='(3, 90)',xy=(3, 90), xytext=(2.5, 100),
            arrowprops=dict(facecolor='black', arrowstyle = '->'))
```
