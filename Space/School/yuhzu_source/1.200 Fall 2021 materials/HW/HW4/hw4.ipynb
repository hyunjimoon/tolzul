```python
import numpy as np
import math
```

```python
def updateR(old, new):
    result = (old + new)/2.0
    return result
def getR(time):
    if time < 30:
        return 3
    elif time < 60:
        return 2
    else:
        return 1
```

```python
e = np.array([0, 0, 0],np.float64)
times = np.array([20, 45, 90],np.float64)
days = 90
print(np.argmax(e))
print(np.random.choice(np.flatnonzero(e == e.max())))
buses = {0:'Express Bus', 1:'Go', 2:'Bunny Bus'}
```

```python
e = np.array([0, 0, 0],np.float64)
for day in range(days):
    if day == 0:
        bus = 2
    else:
        bus = np.random.choice(np.flatnonzero(e == e.max()))
    old = e[bus]
    new = getR(times[bus])
    updatedR = updateR(old, new)
#     print(updatedR)
    e[bus] = updatedR
#     print(e)
    print(f'{day+1} & {buses[bus]} & {np.round(e,2)} \\\\')
print(e)
print(np.argmax(e))
```

```python
e = np.array([0, 0, 0],np.float64)
for actualDay in range(days):
    day = actualDay % 6
    if actualDay ==0:
        bus = 2
    elif day < 3:
        bus = np.random.choice(np.flatnonzero(e == e.max()))
    elif day % 3 == 0:
        bus = 2
    elif day % 3 == 1:
        bus = 1
    else:
        bus = 0
    old = e[bus]
    new = getR(times[bus])
    updatedR = updateR(old, new)
#     print(updatedR)
    e[bus] = updatedR
#     print(e)
    print(f'{actualDay+1} & {buses[bus]} & {np.round(e,2)} \\\\')
print(e)
print(np.argmax(e))
```
