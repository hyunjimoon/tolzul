```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import math
import scipy.stats as st
```

```python
rows = 200
data = pd.DataFrame(columns=['EXP', 'DUR', 'CITY', 'CAMP', 'GRSIZE', 'INC', 'EDUC'])
data['id'] = np.arange(0, rows, 1)
```

```python
# trip duration
p = 0.8
data['DUR'] = np.random.choice(2, rows, p=[p, 1-p])
```

```python
# city
p = 0.7
data['CITY'] = np.random.choice(2, rows, p=[p, 1-p])
```

```python
# camp
p = 0.8
data['CAMP'] = np.random.choice(2, rows, p=[p, 1-p])
```

```python
# group size
data['GRSIZE'] = np.random.choice(np.arange(1,10,1), rows)
```

```python
# income
mu = 25
sigma = 7
data['INC'] = np.random.normal(mu, sigma, rows)
```

```python
data['INC'].min()
```

```python
# education
beta1 = 5 # mean is 5 years if income greater than $25,000
beta2 = 2 # mean is 2 years otherwise
data.loc[data['INC']>25, 'EDUC'] = np.random.exponential(beta1, \
                                                            data[data['INC']>25].shape[0])
data.loc[data['INC']<=25, 'EDUC'] = np.random.exponential(beta2, \
                                                            data[data['INC']<=25].shape[0])
```

```python
# expenditure
data['EXP'] = 20 - 5*data['DUR'] + 15*data['CITY'] - 10*data['CAMP'] - 2.5*data['GRSIZE'] + \
                1.5*data['INC'] + 2.5*data['EDUC'] + np.random.normal(0, 5, rows)
```

```python
# data.groupby('GRSIZE')['id'].nunique()
# data['EDUC'].mean()
data
```

```python
data[['EXP', 'INC', 'EDUC']] = data[['EXP', 'INC', 'EDUC']].astype('float64')
data.dtypes
```

```python
# sample = data.loc[0:99, :].copy()
# sample = data.copy()
sample = data.loc[data['INC']>25, :].copy()
sample
```

```python
stats = sample[['EXP', 'DUR', 'INC', 'EDUC']].describe()
stats
```

```python
np.set_printoptions(suppress=True)
cov = np.cov(sample[['EXP', 'DUR', 'INC', 'EDUC']], rowvar=False)
cov = np.round(cov, 2)
cov
```

```python
# histogram

fig, axs = plt.subplots(2, 2, figsize=(15, 15))

n_bins = 10

cols = ['EXP', 'DUR', 'INC', 'EDUC']
units = ['$ per person', '1 or 0', '$000s', 'Years']

c_count = 0

fontsize = 24

for c in cols:
    axs[int(c_count/2), c_count%2].hist(sample[c], bins=n_bins)
    
    axs[int(c_count/2), c_count%2].set_xlabel(f'{c} ({units[c_count]})', fontsize=fontsize)
    
#     axs[int(c_count/2), c_count%2].set_ylim([0, 1])
    axs[int(c_count/2), c_count%2].set_ylabel('Frequency', fontsize=fontsize)
    
    c_count+=1
```

```python
# z value
delta = stats.loc['mean', 'INC']
sigma = stats.loc['std', 'INC']
n = stats.loc['count', 'INC']
z = delta/(sigma*math.sqrt(n))
z
```

```python
area = st.norm.cdf(z)
area
```
