```python
import numpy as np
import pandas as pd
import random

from mxlMcmc import estimate

import h5py

import matplotlib
import matplotlib.pyplot as plt
from math import floor
```

# Load data


```python
data = pd.read_csv('swissmetro_long.csv')
```

Observe that the data are in long format.

```python
data.head(20)
```

```python
data = data[((data['PURPOSE'] != 1) & (data['PURPOSE'] != 3)) != True]
```

# Prepare data

```python
data.shape
```

```python
indID = np.array(data['ID'].values, dtype = 'int64').reshape((-1,1))
obsID = np.array(data['obsID'].values, dtype = 'int64').reshape((-1,1))
altID = np.array(data['altID'].values, dtype = 'int64').reshape((-1,1))

chosen = np.array(data['chosen'].values, dtype = 'int64')

tt = np.array(data['TT'].values, dtype = 'float64').reshape((-1,1)) / 1
cost = np.array(data['CO'].values, dtype = 'float64').reshape((-1,1)) / 1
he = np.array(data['HE'].values, dtype = 'float64').reshape((-1,1)) / 1
ga = np.array(data['GA'].values, dtype = 'int64').reshape((-1,1))
cost[(altID[:,0] <= 2) & (ga[:,0] == 1),0] = 0

const2 = 1 * (altID == 2)
const3 = 1 * (altID == 3)
```

# Estimate model I
To begin, we estimate a simple MNL model.

```python
xFix = np.hstack((const2, const3, cost / 100, tt / 100))
xRnd = np.zeros((0, 0))

#Fixed parameter distributions
#0: normal
#1: log-normal (to assure that fixed parameter is striclty negative or positive)
xFix_trans = np.array([0, 0, 0, 0])

#Random parameter distributions
#0: normal
#1: log-normal
#2: S_B
xRnd_trans = np.array([0, 0, 0])

paramFix_inits = np.zeros((xFix.shape[1],))
zeta_inits = np.zeros((xRnd.shape[1],))
Omega_inits = 0.1 * np.eye(xRnd.shape[1])

A = 1.04
nu = 2
diagCov = True

mcmc_nChain = 2
mcmc_iterBurn = 20000
mcmc_iterSample = 20000
mcmc_thin = 5
mcmc_iterMem = 20000
mcmc_disp = 1000
seed = 4711
simDraws = 10000

rho = 0.1
rhoF = 0.01

modelName = 'model'
deleteDraws = False

results = estimate(
        mcmc_nChain, mcmc_iterBurn, mcmc_iterSample, mcmc_thin, mcmc_iterMem, mcmc_disp, 
        seed, simDraws,
        rhoF, rho,
        modelName, deleteDraws,
        A, nu, diagCov,
        paramFix_inits, zeta_inits, Omega_inits,
        indID, obsID, altID, chosen,
        xFix, xRnd,
        xFix_trans, xRnd_trans)
```

```python
results.keys()
```

# Estimate model II
Now, we allow some parameters to vary randomly across individuals.

```python
xFix = np.hstack((const2, const3))
xRnd = np.hstack((-cost / 10, tt / 10))

#Fixed parameter distributions
#0: normal
#1: log-normal (to assure that fixed parameter is striclty negative or positive)
xFix_trans = np.array([0, 0])

#Random parameter distributions
#0: normal
#1: log-normal
#2: S_B
xRnd_trans = np.array([0, 0])

paramFix_inits = np.zeros((xFix.shape[1],))
zeta_inits = np.zeros((xRnd.shape[1],))
Omega_inits = 0.1 * np.eye(xRnd.shape[1])

A = 1.04
nu = 2
diagCov = True

mcmc_nChain = 2
mcmc_iterBurn = 20000
mcmc_iterSample = 20000
mcmc_thin = 5
mcmc_iterMem = 20000
mcmc_disp = 1000
seed = 4711
simDraws = 10000

rho = 0.1
rhoF = 0.01

modelName = 'model'
deleteDraws = False

results = estimate(
        mcmc_nChain, mcmc_iterBurn, mcmc_iterSample, mcmc_thin, mcmc_iterMem, mcmc_disp, 
        seed, simDraws,
        rhoF, rho,
        modelName, deleteDraws,
        A, nu, diagCov,
        paramFix_inits, zeta_inits, Omega_inits,
        indID, obsID, altID, chosen,
       xFix, xRnd,
        xFix_trans, xRnd_trans)
```

Let us plot the Markov Chains and check if they  are stationary

```python
paramName = 'zeta'
nParam  = 2
mcmc_iterSampleThin = floor(mcmc_iterSample / mcmc_thin)

postDraws = np.zeros((mcmc_nChain, mcmc_iterSampleThin, nParam))
for c in range(mcmc_nChain):
    file = h5py.File(modelName + '_draws_chain' + str(c + 1) + '.hdf5', 'r')
    postDraws[c,:,:] = np.array(file[paramName + '_store']).reshape((mcmc_iterSampleThin, nParam))


#Trace plots
fig, ax = plt.subplots(1,1,figsize=(6.4, 4.8 / 2))
ax.plot(postDraws[0,:,1], label = 'Chain 1')
ax.plot(postDraws[1,:,1], label = 'Chain 2')

ax.set(xlabel = 'Iteration after burn in', ylabel = r'$\beta$',
       title= r'Trace plot of posterior draws of $\beta$')
plt.legend()
ax.grid()

#fig.savefig("figure_trace.pdf", dpi = 300)
plt.show()


#Posterior
postDist = np.array(postDraws[:,:,1]).reshape((-1,))

fig, ax = plt.subplots(1,1,figsize=(6.4, 4.8 / 2))

num_bins = 50
n, bins, patches = ax.hist(postDist, num_bins, density = True)
ax.set(xlabel = r'$\beta$', ylabel = 'Posterior density',
       title= r'Histogram of posterior draws of $\beta$')

#fig.savefig("figure_hist.pdf", dpi = 300)
plt.show()
```
