```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def  generate_cumm_diagram(time,arrival,cycle_time):
    fig  =  plt.figure()
    ax  =  fig.gca()
    ax.set_xticks(np.arange(0,  50,  1))  #control  x  ticks
    ax.set_yticks(np.arange(0,  10,  1))  #control  y  ticks
    plt.xlim([0,50])  #time
    plt.ylim([0,10])  #space
    fig.set_figheight(10)
    fig.set_figwidth(20)

    time  =  time
    arrival  =  arrival
    cycle_time  =  cycle_time

    exit_time  =  [time[0]  +  cycle_time[0]]

    n  =  sum(arrival)

    y  =  [0]
    x  =  [0]

    for  i  in  range(len(time)):
        x.append(time[i])
        y.append(y[-1])
        x.append(time[i])
        y.append(y[-1]+arrival[i])
        if  n  >  1  :
            print(exit_time)
            if  (  (exit_time[-1]  <=  time[i+1])  ):
                print("A")
                exit_time.append(cycle_time[i+1]+time[i+1])
            else  :
                print("B")
                exit_time.append(exit_time[-1]+cycle_time[i+1])
            
            n=n-1

    y1=  [0]
    x1  =  [0]

    for  i  in  range(len(exit_time)):
        x1.append(exit_time[i])
        y1.append(y1[-1])
        x1.append(exit_time[i])
        y1.append(y1[-1]+1)

    """
    print(x)
    print(y)
    print(exit_time)
    print(x1)
    print(y1)
    """

    plt.plot(x,  y,  label  =  "A(t)",  linestyle="-",linewidth=2,  color  =  'r',  marker  =  'o')
    plt.plot(x1,  y1,  label  =  "D(t)",  linestyle=":",linewidth=2,  color  =  'g',  marker  =  'o')

    plt.xlabel("Time")
    plt.ylabel("Cummulative  N")
    plt.grid()
    plt.legend()
    plt.show()

time  =  [5,10,15]  #  time  of  arrival
arrival = [1,2,1] # arrival number for that time slot
cycle_time = [5,4,3,5] # for sequence in arrvival what is the service time. Note that sum(arrival) = number of terms in cycle time

generate_cumm_diagram(time,arrival,cycle_time)
```
